```python
from datetime import datetime, timedelta
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
headers = {
    "Referer": "https://www.amazon.com/",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8",
    "Accept-Encoding": "gzip, deflate, br",
    "Accept-Language": "en-US,en;q=0.9",
    "Connection": "keep-alive",
    "Upgrade-Insecure-Requests": "1"
}

base_url = f"https://www.amazon.com/s?k=data+engineering+books"
books = []
url = f"{base_url}&page={page}"
response = requests.get(url, headers=headers)
```

```python
page = 10
responses = []
for i in range(10):
    page = i
    url = f"{base_url}&page={page}"
    responses.append(requests.get(url, headers=headers))
```

```python
for response in responses:    
    soup = BeautifulSoup(response.content, "html.parser")
    book_containers = soup.find_all('div', {'data-component-type': 's-search-result'})
    for book in book_containers:
        title_tag = book.find('h2')
        title = title_tag.get_text(strip=True) if title_tag else None
        author_tag = book.find('div', class_='a-row a-size-base a-color-secondary')
        author = author_tag.get_text(strip=True).replace('by', '') if author_tag else None
        rating_tag = book.find('span', class_='a-icon-alt')
        rating = rating_tag.get_text(strip=True) if rating_tag else None

        books.append({
            "Title": title,
            "Author": author,
            "Rating": rating,
        })
```

```python
# Remove duplicates
unique_books = {book['Title']: book for book in books}.values()
unique_books_list = list(unique_books)

print(f"Fetched {len(unique_books_list)} unique books")

df = pd.DataFrame(unique_books_list)
df
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.parse import urljoin
import time

# --- Step 1: Fetch the Webpage ---
# Define the URL of the site you want to scrape


# Use requests.get() to fetch the HTML content
# We add headers to mimic a real browser
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

df = {
    "quote": [],  
    "author": [],
    "tag": []
}
page = 1
while True:
    try:
        url = f"http://quotes.toscrape.com/page/{page}/"
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # This will raise an error for bad responses (like 404 or 500)

        if response.status_code == 404:
            print(f"Page {page} doesn't exist. Stopping.")
            break
        
        # --- Step 2: Parse the HTML ---
        # Create a BeautifulSoup object
        # 'html.parser' is the built-in Python parser
        soup = BeautifulSoup(response.text, 'html.parser')
        
        quotes = soup.find_all("div", class_="quote")
        
        for quote in quotes:
            text = quote.find("span", class_="text").get_text()
            author = quote.find(class_="author").get_text()
            tags = quote.find_all("a", class_="tag")
            arr_tags = []
            for tag in tags:
                arr_tags.append(tag.get_text())
            
            df["quote"].append(text.replace("“", "").replace("”", ""))  # Remove both opening and closing quotes
            df["author"].append(author)
            df["tag"].append(", ".join(arr_tags))

            page+=1
            time.sleep(1)
    
        

        
    except requests.exceptions.RequestException as e:
        print(f"Error fetching URL: {e}")

# Create DataFrame AFTER the loop
df = pd.DataFrame(df)
df

# Optional: Save to CSV
# df.to_csv('quotes.csv', index=False)
```
